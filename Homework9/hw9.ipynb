# 2 Plotting a Function
Goal: Write a **function** that will plot $f(x) = x^2$ and $g(x) = \sqrt{x}$ on the same plot and return the $f(x)$ and $g(x)$ values as separate lists. Your plot needs to have the following:
- An appropriate figure size (`figsize=(10,8)` is recommended)
- Labeled x and y axis
- A title
- Different colors and linestyles for each function
- A legend
- Your domain should be $x \in [0, 10]$. *Hint: Use `np.linspace` to set a range of values for your functions.*

Your function should return:
- A list of $f(x)$ values
- A list of $g(x)$ values.

```python
# Your code here
def plot_functions():
    #Graphs two functions, f(x) = x^2 and g(x) = sqrt(x) against eachother.    x = np.linspace(0, 10, 500)  
    f_x = x**2
    g_x = np.sqrt(x)
    
    plt.figure(figsize=(10, 8))
    plt.title("Plot of $f(x) = x^2$ and $g(x) = \sqrt{x}$", fontsize=16)

    # Plot f(x) = x^2 with a blue solid line, and g(x) = sqrt(x) with a red dashed line
    plt.plot(x, f_x, color='blue', linestyle='-', label=r'$f(x) = x^2$')
    plt.plot(x, g_x, color='red', linestyle='--', label=r'$g(x) = \sqrt{x}$')

    plt.xlabel("x", fontsize=14)
    plt.ylabel("y", fontsize=14)
    
    plt.legend()
    plt.grid(True)

    plt.show()
    
    return f_x.tolist(), g_x.tolist()

plot_functions()
```

```python
# Your code here
def plot_functions():
    #Graphs two functions, f(x) = x^2 and g(x) = sqrt(x) against eachother. Also adds a vertical line, and horizointal line. 
    x = np.linspace(0, 100, 500)  
    
    f_x = x**2
    g_x = np.sqrt(x)
    
    plt.figure(figsize=(10, 8))
    plt.title("Plot of $f(x) = x^2$ and $g(x) = \sqrt{x}$", fontsize=16)

    # Plot f(x) = x^2 with a blue solid line, and g(x) = sqrt(x) with a red dashed line
    plt.plot(x, f_x, color='blue', linestyle='-', label=r'$f(x) = x^2$')
    plt.plot(x, g_x, color='red', linestyle='--', label=r'$g(x) = \sqrt{x}$')
    plt.plot([5, 5], [0, 150], color='green', linestyle=':', label='Vertical Line at x=5')
    plt.plot([0, 12], [50, 50], color='grey', linestyle=':', label='Horizontal Line at y=50')
    
    #Changing limits
    plt.xlim(0, 12) 
    plt.ylim(0, 150)

    plt.xlabel("x", fontsize=14)
    plt.ylabel("y", fontsize=14)
    
    plt.legend()
    plt.grid(True)
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

    plt.show()
    
    return f_x.tolist() + g_x.tolist()

plot_functions()
```
